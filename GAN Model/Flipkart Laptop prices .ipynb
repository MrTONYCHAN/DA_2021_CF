```python
import csv
from bs4 import BeautifulSoup
from selenium import webdriver
import requests 
```

```python
driver=webdriver.Chrome()
```

```python
url="https://www.flipkart.com"
driver.get(url)
```

```python
templates=[]
def get_url(search_term):
    for i in range(1,31):
        template="https://www.flipkart.com/search?q={}&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&page={}"
        search_term=search_term.replace(' ',"%20")
        template=template.format(search_term,i)
        templates.append(template)
```

```python
url=get_url("laptop")
len(templates)
#https://www.flipkart.com/search?q=laptop&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&page=1
```

```python
templates
```

```python
products=[]
prices=[]
ratings=[]
```

```python
for url in templates:
    r=requests.get(url)
    htmlContent=r.content
    soup=BeautifulSoup(htmlContent,"html.parser")
    #soup=BeautifulSoup(driver.page_source,'html.parser')
    #content=driver.page_source.encode("utf-8").strip()
    #soup=BeautifulSoup(content,"html.parser")
    for a in soup.findAll('a',href=True,attrs={'class':'_1fQZEK'}):
        try :
            name=a.find('div',attrs={'class':'_4rR01T'})
            products.append(name.text)
        except:
            products.append("NA")
        try:
            rating=a.find('div',attrs={'class':'_3LWZlK'})
            ratings.append(rating.text)
        except:
            ratings.append('NA')
        try:
            price=a.find('div',attrs={'class':'_30jeq3 _1_WHN1'})
            prices.append(price.text)
        except:
            prices.append("NA")
            
```

```python
import pandas as pd
df=pd.DataFrame({"Product Name":products,
                  "Rating":ratings,
                   "Prices in Rs":prices})
df["Prices in Rs"]=df["Prices in Rs"].str.replace("â‚¹"," ")
```
