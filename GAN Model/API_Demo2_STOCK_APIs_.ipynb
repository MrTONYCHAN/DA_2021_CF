```python
# Be sure to pip install websocket-client
# Details: https://pypi.org/project/websocket-client/
!pip install websocket-client
import websocket

def on_message(ws, message):
    print(message)

def on_error(ws, error):
    print(error)

def on_close(ws):
    print("### closed ###")

def on_open(ws):
    ws.send('{"action":"auth","params":"Authorization: Bearer jBEOJzz9ROsx33prmWSr_lBIYK4tHj8v"}')
    ws.send('{"action":"subscribe","params":"C.AUD/USD,C.USD/EUR,C.USD/JPY"}')

if __name__ == "__main__":
    # websocket.enableTrace(True)
    ws = websocket.WebSocketApp("wss://socket.polygon.io/forex",
                              on_message = on_message,
                              on_error = on_error,
                              on_close = on_close)
    ws.on_open = on_open
    ws.run_forever()
```

#Dataframe

```python
import pandas as pd

df = pd.DataFrame(data['results'])
df
```

#Transposed the Data - all financial statements lines are in the column 

```python
statement = df.T
statement
```

#Manipulation 

Just put the period in the column header such that it will be, very easy to scoop out this yearly annualized data from this data.

```python
statement.columns = list(statement.loc['period'])
statement
```

So, here we are more interested : Yearly data = YA

#Yearly Data(YA)

```python
#  Yearly data = YA

ann_state = statement['YA']
ann_state_fin = ann_state[6:]
ann_state_fin.columns = list(ann_state.loc['calendarDate'])
ann_state_fin
```

# Quaterly Data(Q)

```python
q_state = statement['Q']
q_state_fin = q_state[6:]
q_state_fin.columns = list(q_state.loc['calendarDate'])
q_state_fin
```
